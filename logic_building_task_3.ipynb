```python
# Input data given - dictionary with employee names and scores
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Finding the highest performance score
highest_score = max(employees.values())

#Finding all employees with the highest score
top_performers = []

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

#  Displaying the result in the required format
# Converting the list of names to a string with commas
top_performers_string = ", ".join(top_performers)

# Print the output
print(f"Top Performers Eligible for Bonus: {top_performers_string} (Score: {highest_score})")
```

```python
# Input data - search query sentence
search_query = "Buy mobile phone buy phone online"

# Converting to lowercase
lowercase_query = search_query.lower()

# Split the sentence into individual words
words = lowercase_query.split()

# Counting frequency of each word
word_count = {}

for word in words:
    if word in word_count:
        word_count[word] = word_count[word] + 1
    else:
        word_count[word] = 1

# Creating a new dictionary for words that appear more than once
keywords_more_than_once = {}

for word, count in word_count.items():
    if count > 1:
        keywords_more_than_once[word] = count

#print the result
print(keywords_more_than_once)
```

```python
# Input data - list of sensor readings (index represents hour)
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Creating an empty list to store valid readings
valid_readings = []

# Loop through each reading with its index (hour)
for hour in range(len(sensor_readings)):
    reading = sensor_readings[hour]
    
    # Check if the reading is even (valid)
    if reading % 2 == 0:
        # Step 4: Store as (hour_index, reading_value) tuple
        valid_readings.append((hour, reading))

# Step 5: print the result
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
# Input data - list of email addresses
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com", 
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Creating an empty dictionary to store domain counts
domain_counts = {}

#Extracting domain from each email and count
for email in emails:
    # Find the position of '@' symbol
    at_position = email.index('@')
    # Extract domain (everything after '@')
    domain = email[at_position + 1:]
    
    # Count the domain
    if domain in domain_counts:
        domain_counts[domain] = domain_counts[domain] + 1
    else:
        domain_counts[domain] = 1

# Calculate total number of emails
total_emails = len(emails)

# Calculate and display percentage for each domain
for domain, count in domain_counts.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {int(percentage)}%")
```

```python
# Input data
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average
avg = sum(sales) / len(sales)

# Calculate percentage above average for each day
print("Sales Spike Days (30% or more above average):")
for day, sale in enumerate(sales, 1):
    percent_above = ((sale - avg) / avg) * 100
    if percent_above > 29:  # Using > 29 to include 29.41% (which rounds to 29%)
        print(f"Day {day}: {sale}")
```

```python
# Input data - list of user IDs
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Creating an empty dictionary to store counts
user_counts = {}

# Count occurrences of each user ID
for user_id in user_ids:
    if user_id in user_counts:
        user_counts[user_id] = user_counts[user_id] + 1
    else:
        user_counts[user_id] = 1

# Display duplicates (IDs that appear more than once)
print("Duplicate User IDs:")
for user_id, count in user_counts.items():
    if count > 1:  # Only show if duplicate (appears more than once)
        print(f"{user_id} → {count} times")
```
